```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
bank['default']=np.where(bank['default'].str.contains("yes"),1,0)
bank['housing']=np.where(bank['housing'].str.contains("yes"),1,0)
bank['loan']=np.where(bank['loan'].str.contains("yes"),1,0)
bank['y']=np.where(bank['y'].str.contains("yes"),1,0)
bank
```

```python
bank.shape
```

```python
bank.info()
```

model building

```python
x=pd.concat([bank.iloc[:,0:10],bank.iloc[:,11:]],axis=1)
y=bank.iloc[:,10]
```

```python
classifier=LogisticRegression()
classifier.fit(x,y)
```

Model predictions

```python
y_pred=classifier.predict(x)
y_pred
```

```python
y_pred_df=pd.DataFrame({'act_y':y,'y_pred_prob':y_pred})
y_pred_df
```

Testing model accuracy

```python
confusion_matrix = confusion_matrix(y,y_pred)
confusion_matrix
```

```python
classifier.predict_proba(x)[:,1]
```

```python
fpr,tpr,thresholds=roc_curve(y,classifier.predict_proba(x)[:,1])
plt.plot(fpr,tpr,color='red')
auc=roc_auc_score(y,y_pred)

plt.plot(fpr,tpr,color='red',label='logit model(area  = %0.2f)'%auc)
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.show()

print('auc accuracy:',auc)
```
